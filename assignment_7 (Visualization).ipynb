```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("weather.csv")
df.head(50)
```

```python
df.shape
```

```python
df.plot(kind='scatter',x='WindSpeed9am',y='WindSpeed3pm')
plt.show()
```

```python
plt.plot(['MinTemp','MaxTemp','Rainfall','Evaporation'])
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.show()
```

```python
df.plot(kind='scatter',x = 'MinTemp', y = 'MaxTemp')
plt.show()
plt.hist(x='Humidity3pm',data=df,bins=20)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue='WindGustDir',size=10).map(plt.scatter,'Humidity9am','Pressure9am').add_legend()
plt.show()
```

```python
sns.stripplot(x='RainToday',y='RISK_MM',data=df)
```

```python
sns.countplot(x='WindDir3pm',data =df)
```

```python
df[['Rainfall','Evaporation' ,'Sunshine','WindGustSpeed']].plot(figsize=(15, 15))
plt.show()
```

```python
df[['Humidity9am', 'Humidity3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']].plot(figsize=(20, 20))
plt.show()
```

```python
sns.barplot(x = 'Cloud9am', y = 'Cloud3pm', data=df)
```

```python
sns.violinplot(x='WindGustDir',y='WindGustSpeed',data=df,size=15)
```

```python
sns.violinplot(x='RainTomorrow',y='Rainfall',data=df,size=10)
```

```python
sns.swarmplot(x='Temp9am',y='Temp3pm',data=df)
```
